```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the models
models = ['ARIMA', 'LSTM', 'GRU']
mse = [0.0001, 0.00006, 0.00004]
mae = [0.004, 0.00593, 0.00482]
r_squared = [0.996, 0.99428, 0.99459]

# Create a figure and set of subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Bar graph for MSE
axes[0].bar(models, mse, color='lightblue')
axes[0].set_title('MSE Comparison')
axes[0].set_ylabel('MSE')
axes[0].set_ylim(0, 0.001)
axes[0].set_xticklabels(models, rotation=45, ha='right')

# Bar graph for MAE
axes[1].bar(models, mae, color='lightgreen')
axes[1].set_title('MAE Comparison')
axes[1].set_ylabel('MAE')
axes[1].set_xticklabels(models, rotation=45, ha='right')

# Bar graph for R-Squared
axes[2].bar(models, r_squared, color='lightcoral')
axes[2].set_title('R-Squared Comparison')
axes[2].set_ylabel('R-Squared')
axes[2].set_xticklabels(models, rotation=45, ha='right')
axes[2].set_ylim(0.99, 1.0)

# Tight layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()

```
